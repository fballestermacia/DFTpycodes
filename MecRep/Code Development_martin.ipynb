# Creating the mechanical rep

```python
def chech_permu(permu):
    dim=len(permu[0])
    for i in range(dim):
        sum=0
        for j in range(dim):
            sum=sum+permu[i,j]
        if sum!=1:
            return False
    return True
```

```python
scf='data/P6mmm/fast/non_VDW/phonons/2x2x2/results_scf/CsV3Sb5.scf.pwo'
#scf='data/Fmmm/results_scf/CsV3Sb5.scf.pwi'
#scf='data/Cmcm/results_scf/CsV3Sb5.scf.pwi'

cryst=cell.read_spg(scf)
print(cryst)
cryst=spg.find_primitive(cryst)
print(cryst)
```

```python
sym_data=spg.get_symmetry_dataset(cryst)
rotations=sym_data.get("rotations")
translations=sym_data.get("translations")
```

```python
i=2
sym_prec=4
silent=True

num_atoms=len(cryst[1])
print('# Atoms',num_atoms)
print(rotations[i])
permu=np.zeros([num_atoms,num_atoms])
cells=np.zeros([num_atoms,3])
for j in range(num_atoms):
    new_pos=np.matmul(rotations[i],cryst[1][j])+translations[i]
    if silent==False:
        print(new_pos)
    if np.any(new_pos<0) or np.any(new_pos)>=1:
        for k in range(3):
            while new_pos[k]<0:
                new_pos[k]=new_pos[k]+1
                cells[j,k]=cells[j,k]-1
            while new_pos[k]>=1:
                new_pos[k]=new_pos[k]-1
                cells[j,k]=cells[j,k]+1
    #Find permutation
    for k in range(num_atoms):
        if np.array_equal(np.round(new_pos,decimals=sym_prec),np.round(cryst[1][k],decimals=sym_prec)):
            permu[k,j]=1          #Atom j going to position k
    if silent==False:
        print(new_pos)
        print('***')
if chech_permu(permu):
    print(trs.cryst2cartesian(rotations[i],cryst[0]))
    print(cells)
    print(permu)
```
